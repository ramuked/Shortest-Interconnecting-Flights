Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

Reading the source dataset

```python
data=pd.read_csv('Airports2.csv')
```

```python
data.drop(['Flights','Fly_date','Origin_population','Destination_population','Org_airport_lat','Org_airport_long','Dest_airport_lat','Dest_airport_long'],axis=1)
```

Replacement of Strings

```python
data['Destination_City']=data.Destination_city.str.replace("(,).*","")
```

```python
data.head()
```

```python
data['Origin_City']=data.Origin_city.str.replace("(,).*","")
```

```python
data.head(10)
```

Removing the undesired columns

```python
data.drop(['Flights','Fly_date','Origin_population','Destination_population','Org_airport_lat','Org_airport_long','Dest_airport_lat','Dest_airport_long','Origin_city','Destination_city'],axis=1,inplace=True)
```

```python
data
```

Combining two columns

```python
data['Combined'] = data[['Origin_City', 'Destination_City']].apply(lambda x: ', '.join(x), axis=1)
```

```python
data.head(10)
```

Removing the Duplicates

```python
data.drop_duplicates(subset=['Combined'],inplace=True)
```

```python
data
```

```python
data.drop(['Combined'],axis=1,inplace=True)
```

```python
data.head(5)
```

```python
data.shape
```

Compressing the data by droping extra rows

```python
Final_Data=data.iloc[0:2500]
```

```python
Final_Data
```

```python
Final_Data.drop(['Combined'],axis=1,inplace=True)
```

```python
Final_Data
```

```python
String Upper Case Conversion
```

```python
Final_Data['Origin_City'] = Final_Data['Origin_City'].str.upper()
```

```python
Final_Data['Destination_City']=Final_Data['Destination_City'].str.upper()
```

```python
Final_Data.head()
```

```python
compression_opts = dict(method='zip',
                        archive_name='out.csv')
```

Taking out the processed data as csv

```python
Final_Data.to_csv('out.zip',index=False,compression=compression_opts)
```
